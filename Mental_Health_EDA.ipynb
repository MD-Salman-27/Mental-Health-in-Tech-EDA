# **Project Summary -**
Mental health has become a crucial concern in the technology industry due to high job demands, work pressure, and fast-paced environments. Despite increasing awareness, mental health issues often remain underreported because of stigma, fear of discrimination, and lack of organizational support. This project performs an Exploratory Data Analysis (EDA) on the Mental Health in Tech Survey (2014) dataset to understand employee attitudes, treatment-seeking behavior, and the role of employer support in addressing mental health challenges.

The dataset consists of responses from tech professionals across various countries and includes attributes related to demographics, family history of mental illness, workplace environment, availability of mental health benefits, and openness in discussing mental health issues. Data preprocessing steps such as handling missing values, removing unrealistic age entries, and standardizing categorical variables were performed to ensure data quality and consistency.

Univariate analysis was conducted to study the distribution of key variables such as age, gender, mental health treatment, and employer-provided benefits. Bivariate analysis explored relationships between factors like family history and treatment, workplace interference and mental health conditions, and company policies versus employee openness. Multivariate analysis further examined how multiple workplace factors collectively influence mental health awareness and treatment behavior.

The analysis reveals that mental health issues are prevalent in the tech industry and that employees with supportive workplace policies, mental health benefits, and assured anonymity are more likely to seek treatment. However, the findings also indicate that mental health is still not treated with the same seriousness as physical health in many organizations.

Overall, this project highlights the importance of proactive employer involvement, supportive policies, and open workplace culture in improving mental health outcomes for tech professionals. The insights gained from this EDA can help organizations design better mental health programs and foster a healthier, more inclusive work environment.

# **Problem Statement**-
The technology industry is known for its fast-paced work culture, long working hours, and high performance expectations, which can significantly impact employee mental health. Although mental health awareness has increased in recent years, many employees in the tech sector still hesitate to openly discuss mental health issues or seek treatment due to stigma, fear of negative workplace consequences, and lack of organizational support.

The Mental Health in Tech Survey (2014) dataset contains responses from technology professionals across different regions, capturing information related to demographics, family history of mental illness, mental health treatment, workplace interference, and employer-provided mental health resources. However, raw survey data alone does not clearly reveal the patterns, relationships, and insights required to understand employee behavior and workplace culture regarding mental health.

Therefore, the problem is to analyze and explore this dataset using Exploratory Data Analysis (EDA) techniques to identify trends, relationships, and key factors influencing mental health awareness, treatment-seeking behavior, and employee comfort in discussing mental health issues within the tech workplace. The insights derived from this analysis can help organizations recognize gaps in mental health support and design better policies to improve employee well-being and productivity.

# **Data Wrangling and Preprocessing**

**Load Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
```

**Load the Dataset**

```python
df = pd.read_csv("survey.csv")
```

**Initial Data Inspection**

```python
# View first 5 rows
df.head()
```

```python
# Check shape of dataset
df.shape
```

```python
# Check column names
df.columns
```

**Dataset Information**

```python
# Basic information about dataset
df.info()
```

```python
# Statistical summary (numerical columns)
df.describe()
```

**Check Missing Values**

```python
# Check missing values count
df.isnull().sum()
```

**Handle Missing Values**

```python
# Fill missing values in self_employed
df['self_employed'] = df['self_employed'].fillna('No')

# Fill missing values in work_interfere
df['work_interfere'] = df['work_interfere'].fillna("Don't know")

# Fill missing state values
df['state'] = df['state'].fillna('Not Applicable')
```

**Data Cleaning - Age Column**

```python
# Remove unrealistic age values
df = df[(df['Age'] >= 10) & (df['Age'] <= 80)]
```

**Data Cleaning - Gender Standardization**

```python
# Convert to lowercase
df['Gender'] = df['Gender'].str.lower()

# Standardize gender values
df['Gender'] = df['Gender'].replace({
    'male': 'Male', 'm': 'Male', 'man': 'Male',
    'female': 'Female', 'f': 'Female', 'woman': 'Female'
})

# Replace other values as 'Other'
df.loc[~df['Gender'].isin(['Male', 'Female']), 'Gender'] = 'Other'
```

**Remove Irrelevant Columns**

```python
# Drop unnecessary columns
df.drop(['Timestamp', 'comments'], axis=1, inplace=True)
```

**Check Duplicates**

```python
# Check duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicates if any
df.drop_duplicates(inplace=True)
```

**Final Cleaned Dataset Check**

```python
df.info()
```

```python
df.isnull().sum()
```

# **Data Visualization and Exploratory Analysis**

**Univariate Analysis**

**Age Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Respondents')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**Insight:**
Most respondents belong to the 25â€“40 age group, indicating a young tech workforce.

**Gender Distribution**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

**Insight:**
Male respondents form the majority, highlighting gender imbalance in the tech industry.

**Mental Health Treatment**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='treatment', data=df)
plt.title('Mental Health Treatment Status')
plt.show()
```

**Insight**:
A significant number of respondents have sought treatment for mental health conditions.

**Availability of Mental Health Benefits**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='benefits', data=df)
plt.title('Availability of Mental Health Benefits')
plt.show()
```

**Insight:**
Not all organizations provide mental health benefits, indicating scope for improvement.

**Bivariate Analysis**

**Family History vs Treatment**

```python
plt.figure(figsize=(7,5))
sns.countplot(x='family_history', hue='treatment', data=df)
plt.title('Family History vs Mental Health Treatment')
plt.show()
```

**Insight:**
Employees with a family history of mental illness are more likely to seek treatment.

**Age vs Mental Health Treatment**

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='treatment', y='Age', data=df)
plt.title('Age vs Mental Health Treatment')
plt.xlabel('Treatment')
plt.ylabel('Age')
plt.show()
```

**Insight**:
The age distribution of employees who sought treatment is slightly broader, indicating mental health concerns across age groups.

**Age vs Work Interference**

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='work_interfere', y='Age', data=df)
plt.title('Age vs Work Interference due to Mental Health')
plt.show()
```

**Insight**:
Employees reporting frequent work interference are spread across multiple age ranges.

**Mental Health Benefits vs Treatment**

```python
benefit_treat = pd.crosstab(df['benefits'], df['treatment'])

benefit_treat.plot(kind='bar', stacked=True, figsize=(7,5))
plt.title('Mental Health Benefits vs Treatment')
plt.xlabel('Benefits')
plt.ylabel('Count')
plt.show()
```

**Insight**:
Employees in organizations offering mental health benefits are more likely to seek treatment.

**Treatment vs Willingness to Discuss with Supervisor**

```python
plt.figure(figsize=(7,5))
sns.pointplot(x='treatment', y=df['supervisor'].map({'Yes':1, 'No':0}), data=df)
plt.title('Treatment vs Willingness to Discuss with Supervisor')
plt.ylabel('Willingness (1=Yes, 0=No)')
plt.show()
```

**Insight**:
Employees who have sought treatment show slightly higher openness in discussing mental health with supervisors.

**Work Interference vs Treatment**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='work_interfere', hue='treatment', data=df)
plt.title('Work Interference vs Treatment')
plt.show()
```

**Insight**:
Higher work interference is associated with increased treatment-seeking behavior.

**Benefits vs Willingness to Seek Help**

```python
plt.figure(figsize=(7,5))
sns.countplot(x='benefits', hue='seek_help', data=df)
plt.title('Benefits vs Mental Health Support Resources')
plt.show()
```

**Insight**:
Organizations offering benefits are more likely to provide mental health resources.

**Multivariate Analysis**

**Treatment by Gender and Benefits**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Gender', hue='treatment', data=df)
plt.title('Treatment by Gender')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='benefits', hue='treatment', data=df)
plt.title('Treatment vs Benefits')
plt.show()
```

**Insight**:
Employer-provided benefits positively influence treatment across genders.

**Mental vs Physical Health Perception**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='mental_vs_physical', data=df)
plt.title('Mental Health vs Physical Health Perception')
plt.show()
```

**Insight:**
Mental health is often not treated with the same importance as physical health.

**Correlation Heatmap (Numerical)**

```python
plt.figure(figsize=(8,6))
numeric_df = df[['Age']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# **Conclusion**
This project conducted an in-depth Exploratory Data Analysis (EDA) on the Mental Health in Tech Survey (2014) dataset to understand employee attitudes, treatment behavior, and organizational support related to mental health in the technology sector. Through systematic data cleaning, visualization, and analysis, meaningful patterns and relationships were uncovered.

The analysis revealed that mental health concerns are prevalent among tech professionals, with a significant number of employees reporting mental health conditions that interfere with their work. These insights highlight the need for organizations to create a more inclusive, supportive, and stigma-free work environment.

Overall, this EDA emphasizes the critical role of employer policies and workplace culture in promoting mental well-being.
